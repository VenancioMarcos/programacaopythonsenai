```python
import xlsxwriter as xlsx
import os

caminho = 'C:\\Users\\FIC\\Documents\\Marcos.Venancio-Python\\'
arquivo = 'Exemplo_10.xlsx'
workbook = xlsx.Workbook(caminho+arquivo) #...........Criar uma nova pasta de trabalho.
sheetDados = workbook.add_worksheet('Dados')

dados = [
    ['produto','qtde','preço','subtotal'],
    ['calça',5,30],
    ['camisa',6,30],
    ['sapato',7,30],
    ['saia',8,30],
    ['blusa',9,30],
    ['shorts',4,30]
]

title = workbook.add_format({
                                'bg_color':'blue',
                                'font_color':'white',
                                'font_name': 'Arial',
                                'align': 'center',
                                'bold':True
})

rotulosLinha = workbook.add_format({
                                'font_color':'gray',
                                'font_name': 'Arial',
                                'align': 'center'
                                
})

qtde = workbook.add_format({
                                'font_color':'gray',
                                'font_name': 'Arial',
                                'align': 'center'
                                
})

valores = workbook.add_format({
                                'font_color':'gray',
                                'font_name': 'Arial',
                                'align': 'center',
                                'num_format':'R$ * #.##,0'
                                
})

formatDatas = workbook.add_format({
                                'font_color':'gray',
                                'font_name': 'Arial',
                                'align': 'center',
                                'num_format':'dd/mm/yyyy'

})

formatHoras = workbook.add_format({
                                'font_color':'gray',
                                'font_name': 'Arial',
                                'align': 'center',
                                'num_format':'h:mm'

})

Mesclado = workbook.add_format({
                                'font_color':'black',
                                'font_name': 'Arial',
                                'font_size':22,
                                'align': 'center',
                                'bold':True

})

sheetDados.merge_range('A1:D1','Planilha de Solicitação', Mesclado)

sheetDados.write('A3','Data')
sheetDados.write_formula('B3', '=Today()', formatDatas )
sheetDados.write('A4','Horas')
sheetDados.write_formula('B4', '=Now()', formatHoras )



posLinha = 5
posColuna = 0

for linha, listas in enumerate(dados):
    if linha == 0:
        sheetDados.write_row(linha + posLinha,posColuna,listas,title)
    
    else:
        sheetDados.write(linha + posLinha,posColuna,listas[0])
        sheetDados.write(linha + posLinha,posColuna+1,listas[1])
        sheetDados.write(linha + posLinha,posColuna+2,listas[2])
        sheetDados.write_formula(linha + posLinha,posColuna + 3 ,'=B' + str(linha + posLinha + 1) + '*C' + str(linha + posLinha + 1))

workbook.close()

os.startfile(caminho + arquivo)
```
