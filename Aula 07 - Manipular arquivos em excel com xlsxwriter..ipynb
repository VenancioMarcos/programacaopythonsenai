### EXERCICIO_01

```python
#Criar e Manipular arquivos no excel

import xlsxwriter as xlsx

#Gerenciar arquivos no sistema operacional

import os

#gerar caminho e um arquivo

caminho = 'C:\\Users\\FIC\\Documents\\Marcos.Venancio-Python\\'
arquivo = 'Exemplo-2.xlsx'

#Criar um documento em excel

workbook = xlsx.Workbook(caminho + arquivo)

sheet = workbook.add_worksheet()

#escrever nas celulas

sheet.write('B1','Salario')
sheet.write('A2','Janaina')
sheet.write('A3','Aysha')
sheet.write('A4','Thomas')
sheet.write('A5','Khevyn')
sheet.write('A6','Marcos')
sheet.write('C1','data')
sheet.write('D1','total')

#fechar e salvar arquivo

workbook.close()


#abrir o documento

os.startfile(caminho + arquivo)
```
